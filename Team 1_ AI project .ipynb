```python
%matplotlib widget
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
car_data = 'processes2.csv'
car_data_df = pd.read_csv('processes2.csv')
car_data_df
```

```python
car_data_df.drop("Unnamed: 0", axis=1 , inplace = True )
car_data_df
```

```python
car_data_df = car_data_df.drop(columns=["Mileage Unit", "Engine (CC)"])
car_data_df
```

```python


car_data_df = car_data_df.rename(
    columns = {
        "name": "Make", "year": "Year", "selling_price": "Price", "km_driven": "Odometer" , "fuel": "Fuel","seller_type": "Seller Type",
    "transmission": "Transmission","owner": "Owner","seats": "Seats","max_power (in bph)": "BPH"
   }
)

car_data_df
```

```python
avg_price_per_make = car_data_df.groupby('Make')['Price'].mean()
```

```python
avg_price_per_make = avg_price_per_make.reset_index()
```

```python
avg_price_per_make.plot(kind='bar', x='Make', y='Price')
```
